```python
# Image Reading using OpenCV2
import torch
import torchvision
from torchvision.io import read_image
from torchvision.utils import draw_bounding_boxes
```

```python
#Read Image
img = read_image('/content/cat.jpg')

# [x1,y1,x2,y2]
# (x1,y1) = top_left
# (x2,y2) = bottom_right

bbox = [290,115,405,385]
bbox=torch.tensor(bbox,dtype=torch.int)
print(bbox)
print(bbox.size())
bbox = bbox.unsqueeze(0)
print(bbox.size())

img = draw_bounding_boxes(img,bbox,width=3,colors = (255,0,0))
img = torchvision.transforms.ToPILImage()(img)
img
```

```python
# multiple images for a bbox without labels
img = read_image('/content/cat_dog.jpg')
bbox1 = =[30,45,330,450]
bbox2 = [320,150,690,460]
bbox = [bbox1,bbox2]
bbox = torch.tensor(bbox,dtype=torch.int)
print(bbox)
print(bbox.size())

img = draw_bounding_boxes(img,bbox,width=3,colors=[(255,0,0),(0,0,255)])
img = torchvision.transforms.ToPILImage()(img)
img
```

```python
# multiple images for a bbox with labels
img = read_image('/content/cat_dog.jpg')
bbox1 = [30,45,330,450]
bbox2 = [320,150,690,460]
bbox = [bbox1,bbox2]
bbox = torch.tensor(bbox,dtype=torch.int)
labels = ['Cat','Dog']
print(bbox)
print(bbox.size())

img = draw_bounding_boxes(img,bbox,width=3,labels = labels,colors=[(255,0,0),(0,0,255)],fill=True,font_size = 24)
img = torchvision.transforms.ToPILImage()(img)
img
```
